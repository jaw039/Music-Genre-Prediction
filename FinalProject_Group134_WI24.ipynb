# 2. Background and Prior Work


Music genre classification is a prominent subject of interest for music streaming platforms (e.g., Spotify, Apple Music, YouTube Music), social media platforms (e.g., TikTok, Instagram, etc.,), as well as advertising companies. It is crucial for various applications, such as personalized music recommendations, content categorization, and enhancing user experiences in streaming platforms. There are various attributes and features of a song that may determine its genre, including acoustic properties, tempo, instrumentalness, and popularity metrics (such as streaming counts or chart rankings).


A study<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) done by Navoda Senavirathne utilizes audio (danceability, energy, key, loudness, mode, speechless, acousticness, instrumentalness, liveness, valence, tempo, duration_ms, time_signature) features scraped using Spotify’s web API to predict the genre (pop, rock, r&b, edm , jazz, blues, country, classical, metal, or hip hop) of a given song. Using mutual information, the author selected the top 80% of features (all features except key, mode, and liveness). They then carried out a correlation analysis, and found that energy has a strong positive correlation with loudness but a strong negative correlation with acousticness, so they drop the energy feature to avoid multicollinearity. This completes feature selection, with the final feature set being: danceability, loudness, speechiness, acousticness, instrumentalness, liveness, valence, tempo, duration_ms, time_signature. After training various machine learning models (Logistic Regression, Decision Tree, Random Forest, XGB Classifier),  Senavirathne found the XGBClassifier to perform the best, with an overall accuracy of 0.647. Interestingly, the pop, r&b, and rock genres had the least accuracy (< 0.5), whereas classical, hip hop, and metal had the highest accuracies (0.896, 0.793, and 0.798, respectively). From this, we can conclude that audio features can be used for genre identification, however, some genres are more easily identifiable than others, and some genres often get mislabeled for other genres.

Kehan Luo’s study<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) uses data from Spotify and Kraggle about top listened to songs to predict the genre of a song based on features including acousticness, key, energy, and tempo. Certain features from Spotify were not included in the study because Luo believed they were not relevant, including time signature, duration time, and valence. A model was trained to return the genre label with the highest probability of being associated with a song given all of the song’s features. Luo concludes that the model, with a final accuracy of 46.9%, is not very reliable and that the variability of certain features within genres as well as the overlap in genres of certain songs made it difficult to accurately predict the genres of songs. However, there were certain distinct differences in genres Luo also pointed out, such as ‘speechiness’ being the best predictor of genre and certain genres having lower or upper limits to key features. In addition to that, Luo believes the model would be significantly more accurate if given more predictions per song. The limitations in Luo’s study, as well as the proposals he gives for further improvements to the model, are helpful in designing future predictive models like his.

In Jie Ren's study.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) on predicting music genres through machine learning, the research delves deep into how audio metadata and textual features can effectively classify music tracks into genres. By rigorously comparing the performance of various machine learning algorithms—specifically Naive Bayes, KNN, Decision Trees, and Random Forest—the study not only identifies the strengths and weaknesses of each method but also places a particular emphasis on Naive Bayes for its remarkable accuracy. This focus highlights the pivotal role of meticulous feature selection and the strategic preprocessing of data, underscoring how these elements critically enhance the model's predictive capabilities. Ren's work meticulously illustrates the sophisticated approach required to accurately categorize music genres, providing valuable insights for future research to further refine and improve model precision in music genre classification.

## 4.2 Data Cleaning

###4.3.3 Comparing Categorical Features Between Genres

Below, we will graph the distribution of categorical variable `mode` between different genres. From this graph, we can see that there are indeed differences in mode proportions between observations of different genres

````python
# Graph distribution of 'mode' between different genres


# Calculate proportion of different mode values for each genre
mode_proportions = df.groupby(['music_genre', 'mode']).size().unstack(fill_value=0)
mode_proportions['Total'] = mode_proportions.sum(axis=1)
mode_proportions = mode_proportions.div(mode_proportions['Total'], axis=0)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
mode_proportions.drop(columns=['Total']).plot(kind='bar', stacked=True, ax=ax)
plt.title('Distribution of Mode for Each Genre')
plt.xlabel('Genre')
plt.ylabel('Proportion')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Mode')
plt.tight_layout()
plt.show()
````

Similary, we will graph the distribution of categorical variable `key` between different genres. The coloring of the bars on the graph matches that of Section 4.3.2 (e.g., gray represents "Rock", brown represents "Rap", etc.). From these distributions, we can see that the 'key' does slighty vary between genres, but the difference is not very stark.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Graph distribution of 'key' between different genres
key_proportions = df.groupby(['music_genre', 'key']).size().unstack(fill_value=0)
key_proportions['Total'] = key_proportions.sum(axis=1)
key_proportions = key_proportions.div(key_proportions['Total'], axis=0)

keys = key_proportions.columns[:-1]  # Exclude 'Total' column
num_keys = len(keys)
num_rows = 4
num_cols = 3

fig, axs = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(18, 14), sharex=True)

for i, key in enumerate(keys):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]
    bars = ax.bar(range(len(key_proportions)), key_proportions[key], color=[genre_color_mapping[genre] for genre in key_proportions.index])
    ax.set_title(f'Distribution of Key {key} for Each Genre')
    ax.set_ylabel('Proportion')
    ax.set_xlabel('Genre')
    ax.set_xticks(range(len(key_proportions)))
    ax.set_xticklabels(key_proportions.index, rotation=45, ha='right')

# Hide unused subplots
for i in range(num_keys, num_rows * num_cols):
    row = i // num_cols
    col = i % num_cols
    axs[row, col].axis('off')

plt.tight_layout()
plt.show()
````

## 7.2 Model Results and Evaluation


#### 7.2.0 Evaluation Metrics

For evaluation, we will use the metrics provided by scikit-learn's `classification_report()`. A description of the metrics is provided below:
* **Precision**: Precision measures the accuracy of positive predictions. It's the ratio (TP) / (TP + FP) for each class, where TP = true positives and FP = false positives. A high precision means a low false positive rate.
* **Recall (Sensitivity)**: Recall is the ability of the classifier to find all positive instances. It's the ratio (TP) / (TP + FN) for each class, where TP = true positives and FN = false negatives. A high recall means a low false negative rate.
* **F1-score**: This metric provides a balanced insight on both precision and recall and is calculated as 2 * (precision * recall) / (precision + recall). F1-score's range from 0 (worst possible value) to 1 (best possible value).
* **Support**: Support is the number of actual occurrences of the class in the testing set. For our dataset, the support for each class is rougly equivalent.
* **Accuracy**: This is the ratio of the number of correctly predicted instances to the total number of instances.
* **Macro avg**: This is the non-weighted average precision, recall, and F1-score across all classes (each class is given equal weight, regardless of size).
* **Weighted avg**: This is weighted average precision, recall, and F1-score across all classes. The weight of each class when calculating the average is its support (the number of true instances).

\
Our proposed model can be improved by better parameter tuning. Currently, we have only tuned the max_iter parameter. However, the logistic regression model in scikit-learn has multiple other parameters that we can also tune using GridSearch to get the optimial parameter combination:

*   penalty: This is the the norm used in the penalization. It can be 'l1' (Lasso) or 'l2' (Ridge). The default value that we used was 'l2', but we can experiment with 'l1'
* C: This is the inverse of regularization strength. Regularization helps prevent overfitting by penalizing large coefficients, so this would be good to tune especially when we have a larger feature set.
* solver: This is the algorithm used. The default was 'lbfgs', but we can also try 'liblinear', 'newton-cg', 'sag', and 'saga'.
* class_weight: These are the weights associated with classes. While our dataset is very balanced, explicitly descibing exact weights may slightly improve performance. By default, all classes are assigned equal weights.
